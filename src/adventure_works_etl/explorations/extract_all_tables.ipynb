```python
spark.sql("USE CATALOG `adventure-work`")
```

```python
spark.sql("CREATE VOLUME IF NOT EXISTS workspace.default.adventure_works_dump")
```

```python
volume_path = "/Volumes/workspace/default/adventure_works_dump"
```

```python
list_tables = [
    "sales.salesorderheader",
    "sales.salesorderdetail",
    "sales.specialoffer",
    "sales.specialofferproduct",
    "sales.currencyrate",
    "sales.salesterritory",
    "production.product",
    "production.productsubcategory",
    "production.productcategory",
    "person.person",
    "person.emailaddress",
    "person.address",
    "person.stateprovince",
    "person.countryregion",
    "sales.customer",
]
print(len(list_tables))
```

```python
for i, table_name in enumerate(list_tables):
    print(f"--------Processing table {i + 1}/{len(list_tables)}: {table_name}")
    (spark.table(table_name).repartition(10).write.format("json").mode("overwrite").save(f"{volume_path}/{table_name}"))
```
